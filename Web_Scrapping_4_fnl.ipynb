# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
# Access chrome browser using webdrive
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)


#Access the website
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(3)


# creating empty dictionary
Most_Viewed = {}
Most_Viewed['Rank'] = []
Most_Viewed['Name'] = []
Most_Viewed['Artist'] = []
Most_Viewed['Upload_date'] = []
Most_Viewed['Views'] = []
time.sleep(3)


a = 1 #counter value


# Scraping the all data
for i in range(0,30):
    time.sleep(3)
    
    try:
        #rank
        rank = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[1]".format(a))
        Most_Viewed['Rank'].append(rank.text)
        time.sleep(2)
            
        #name
        name = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[2]".format(a))
        Most_Viewed['Name'].append(name.text)
        time.sleep(2)
        
        #artist
        artist = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[3]".format(a))
        Most_Viewed['Artist'].append(artist.text)
        time.sleep(2)
    
        #upload_date
        uploadDate = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[5]".format(a))
        Most_Viewed['Upload_date'].append(uploadDate.text)
        time.sleep(2)
    
        #views
        view = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[4]".format(a))
        Most_Viewed['Views'].append(view.text)
        time.sleep(2)
    except NoSuchElementException as n:
        print("Exception",n)
   
    # incresing counter variable
    a += 1


Most_Viewed = pd.DataFrame(Most_Viewed)
Most_Viewed[0:10]
```

Url = https://www.guru99.com/

A) Name

B) Description

Note: - From guru99 home page you have to reach to selenium exception handling page through code

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get('https://www.guru99.com/')

selenium_link=driver.find_element_by_xpath("//ul[@class='menu1'][1]/li[3]")
try:
    selenium_link.click()
except ElementNotInteractableException:          #handling element not clickable exception
    driver.get(selenium_link.get_attribute('href'))
time.sleep(2)

#clicking on Selenium Exception Handling (Common Exceptions List) under Advance Webdriver Stuff!
exception_handle_link=driver.find_element_by_xpath("//a[@title='Selenium Exception Handling (Common Exceptions List)']")
try:
    exception_handle_link.click()
except ElementNotInteractableException:      #handling element not clickable exception
    driver.get(exception_handle_link.get_attribute('href'))              
time.sleep(2)

name = []

try:
    name_tag = driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr/td[1]")
    for i in name_tag:
        name.append(i.text)
except NoSuchElementException:#handling no such element exception
    name.append('No details available')
time.sleep(2)

#scraping exceptions description
desc=[]
try:
    desc_list=driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr/td[2]")
    for i in desc_list[1:]:
        desc.append(i.text)
except NoSuchElementException:#handling no such element exception
    desc.append('No details available')
time.sleep(2)

#creating dataframe
df=pd.DataFrame({"Exception_name":name[0:40],
                "Description":desc[0:40]})
df
```

6. Scrape the details of top 100 songs on billiboard.com.
Url = https:/www.billboard.com/
You have to find the following details:

A) Song name

B) Artist name

C) Last week rank

D) Peak rank

E) Weeks on board


```python
driver = webdriver.Chrome("chromedriver.exe")
url =  "https://www.billboard.com/charts/hot-100"
driver.get(url)

Song_Name =[]
Singer=[]
rank=[]
Last_Week=[]
Weeks_on_board=[]

#scraping the Song_Name 
son=driver.find_elements_by_xpath("//li[@Class='lrv-u-width-100p']")
for i in son:
    if i.text is None :
        Song_Name.append("--") 
        Singer.append("--") 
        rank.append("--") 
        Last_Week.append("--") 
        Weeks_on_board.append("--") 
    else:
        Song_Name.append(i.text.split('\n')[0])
        Singer.append(i.text.split('\n')[1])
        rank.append(i.text.split('\n')[2])
        Last_Week.append(i.text.split('\n')[3])
        Weeks_on_board.append(i.text.split('\n')[4])


BillBOARD_HOT100=pd.DataFrame([])
BillBOARD_HOT100['Song_Name']=Song_Name
BillBOARD_HOT100['Peak_rank']=rank
BillBOARD_HOT100['Singer / Crew']=Singer
BillBOARD_HOT100['Last_Week_Rank']=Last_Week
BillBOARD_HOT100['Weeks_on_board']=Weeks_on_board

BillBOARD_HOT100
```

B) Year span

C) Genre

D) Run time

E) Ratings

```python
driver=webdriver.Chrome("chromedriver.exe") 
url ="https://www.imdb.com/list/ls095964455/"
driver.get(url)
driver.maximize_window()

# Creating empty list
Name=[]
Year_span=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]

#Scraping Name
name=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
for i in name:
    Name.append(i.text)
    
#Scraping Year_span
year_span=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]")
for i in year_span:
    Year_span.append(i.text)

    
#Scraping Genre
genre=driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[5]")
for i in genre:
    Genre.append(i.text)
    
    
#Scraping Run_time
run_time=driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[3]")
for i in run_time:
    Run_time.append(i.text)
    
#Scraping Ratings
ratings=driver.find_elements_by_xpath("//div[@class='ipl-rating-widget']/div/span[2]")
for i in ratings:
    Ratings.append(i.text)
    
#Scraping Votes
votes=driver.find_elements_by_xpath("//p[@class='text-muted text-small'][3]/span[2]")
for i in votes:
    Votes.append(i.text)
    
Top_shows=pd.DataFrame({})
Top_shows['Name']=Name
Top_shows['Year_span']=Year_span
Top_shows['Genre']=Genre
Top_shows['Run_time']=Run_time
Top_shows['Ratings']=Ratings
Top_shows['Votes']=Votes
Top_shows
```
